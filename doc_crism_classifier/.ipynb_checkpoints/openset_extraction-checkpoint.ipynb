```python
import numpy as np
from matplotlib import pyplot as plt
import os
import sys
sys.path.append("/Volume2/arunFiles/CRISM_minMapping_Ident/python_codeFiles/")

from doc_crism_classifier import doc_crism_classifier
from hsiUtilities.crism_ganProc_utils import crism_ganProc_utils

%matplotlib inline
```

```python
'Load the labeled data'
data = np.load('data.npz')
x_train, x_test = data['name1'], data['name2']
y_train, y_test = data['name3'], data['name4']

'Estimate the number of classes'
n_classes = max(max(y_train), max(y_test)) + 1

'Get the unlabeled data'
data_open = np.load('data_open.npz')
x_openset, y_openset = data_open['name1'], data_open['name2']

'Combine the labeled test data and the unlabeled data'
x_full, y_full = np.vstack((x_test, x_openset)), np.hstack((y_test, y_openset))
```

```python
'Create the feature extractor of interest'
dis_rep = crism_ganProc_utils().create_rep_model()
    
'Create and load the best DOC model'
obj1 = doc_crism_classifier(int(n_classes), disrep_model=dis_rep, model_name='doc_ganRep_crismData')
    
'Load a pretrained model'
obj1.load_preTrained_model(model_loc=os.path.join('/Volume2/arunFiles/CRISM_minMapping_Ident/python_codeFiles',
                                                     'doc_crism_classifier/doc_ganRep_crismData-1/doc_model'))
```

```python
'Get the prediction of the DOC on the testdata'
y_testpred = np.asarray(obj1.predict_test(x_openset, scale=0.9))
```

```python
idx_4 = np.where(y_testpred == 4)[0]
idx_0 = np.where(y_testpred == 7)[0]
idx_3 = np.where(y_testpred == 3)[0]
idx_8 = np.where(y_testpred == 8)[0]
```

```python
fig1, axs = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(7, 7))
axs[0][0].plot((x_openset[idx_4[:10], :]).T)
axs[0][0].set_title('EM-4(Alunite) outside ident');
axs[0][1].plot((x_openset[idx_0[:10], :]).T)
axs[0][1].set_title('EM-0(Olivine) outside ident');
axs[1][1].plot((x_openset[idx_3[:10], :]).T)
axs[1][1].set_title('EM-3(MHS) outside ident');
axs[1][0].plot((x_openset[idx_8[:10], :]).T)
axs[1][0].set_title('EM-8(Al-Smect) outside ident');
```
